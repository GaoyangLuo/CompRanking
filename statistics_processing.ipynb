```python
df_hmm=pd.read_csv("./test/CompRanking/Virulence/hmm_result/ERR1191817.contigs_5M_contigs_hmm.csv", 
                  sep='\t', header=0, engine="python")
df_hmm
```

```python
positive_query=pd.read_csv("./databases/models_and_domains/positive_domains.tsv", 
                  sep='\t', header=None, engine="python")
positive_query
```

```python
positive_query_1=positive_query.copy()
positive_query_1["class"]="pathogenic"
positive_query_1.columns=["query_positive","class"]
positive_query_1
```

```python
# positive_merge_tmp=pd.merge(positive_query_1,shared_query_1,how="left")
# positive_merge_tmp
positive_merge_tmp=pd.merge(df_hmm,positive_query_1,left_on="query", right_on="query_positive",how="inner")
positive_merge_tmp
```

```python
#获得去重的ORF以及对应的毒性
Virulence_non_redundant=positive_merge_tmp.loc[:,["ID","class"]]
Virulence_non_redundant
```

```python
#unique
Virulence_non_redundant_unique=Virulence_non_redundant.drop_duplicates()
Virulence_non_redundant_unique
```

```python
#最终生成orf的VF预测csv
Virulence_non_redundant_unique.to_csv("./test/CompRanking/Virulence/ERR1191817.contigs_5M_contigs_VF_Prediction.csv", sep="\t")
```
