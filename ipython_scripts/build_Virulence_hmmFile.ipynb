```python
#open VFDB summary file
df_VFs_meta=pd.read_csv("/lomi_home/gaoyang/software/CompRanking/databases/VFDB/VFs.csv",sep=",",header=0,engine="python")
df_VFs_meta
```

```python
#extract key columns
df_VFs_meta=df_VFs_meta[["VFID","VF_Name","Bacteria","VFCID","VFcategory"]]
df_VFs_meta["ESKAPE"]="-"
df_VFs_meta
```

```python
#find ESKAPE
ESKAPE_list=['Enterococcus faecium',
'Staphylococcus aureus',
'Klebsiella pneumoniae',
'Acinetobacter baumannii',
'Pseudomonas aeruginosa', 
'Enterobacter']

for index, name in df_VFs_meta.iterrows():
    if df_VFs_meta["Bacteria"][index] in ESKAPE_list:
        df_VFs_meta["ESKAPE"][index] = "Positive"
    else:
        df_VFs_meta["ESKAPE"][index] = "Negative"
df_VFs_meta
```

```python
#extract index ">" and descrition of VFDB_setA_pro.fas
def extractVFinfo_setA(f):
	info_head = [] 
	for i in f: #open VFDB_setA_pro.fas
		if re.match(">",i):
			info_head.append(i)
	return info_head
#extract ">"index
# def cutoff(header_list):
#     dicVFID={}
    

f=open("/lomi_home/gaoyang/software/CompRanking/databases/VFDB/VFDB_setA_pro.fas","r")

header_list=extractVFinfo_setA(f)

num_string="0123456789"
dicVFID={}
tar = re.compile(r'[(](.*?)[)]', re.S) #find ()
final_check=[]
for i in header_list:
    VFid=[]    
    s=i.split(" ")
    s1_find=re.findall(tar,s[1])
    VF_fasID=s[0].strip(">")
    dicVFID[VF_fasID]=s1_find[0] #write head index 
    
     
    if re.match(">",i): 
        #process content in the ()    
        findall=re.findall(tar, i) #what's in ()
        findall.pop()
        for j in findall:# find VFID
            if j.startswith("VF"):
                VFid.append(j)
                final_check.append(j)

    
    dicVFID[VF_fasID]=dicVFID[VF_fasID] + "+" + VFid[0]
dicVFID
            
#         vfcid=findall.pop(-1)
#         dicVFID[s[0]]=dicVFID[s[0]] +"/"+vfcid
# dicVFID
    
    # # print(s)  #index: 1,6,9
    # if re.match(">",i):       
    #     tar = re.compile(r'[(](.*?)[)]', re.S)
    #     findall=re.findall(tar, i)
    #     findall.pop(0)
    #     findall_str="/".join(findall)
    #     dicVFID[s[0]]=findall_str

#check -> OK
# lenth=[]
# for i in dicVFID:
#     a=dicVFID[i].split("+")
#     b=a[1]
#     if not b.startswith("VF"):
#         print(b)

    
    # print(s[0].strip(">"))
    # dicVFID[s]=""
    
    # if re.match(">",i):
    # dicVFID[i]=""
#     for j in i.split(" "):
#         if re.match(">",j):
#             dicVFID[j]=""        
#     tar = re.compile(r'[(](.*?)[)]', re.S)
#     findall=re.findall(tar, string)
#     dicVFID[j]= dicVFID[j]+ findall

# print(dicVFID)     
```

```python
df_vfdic = pd.DataFrame(pd.Series(dicVFID),columns=["tmp"])
df_vfdic["fasID"]="-"
df_vfdic["VFID"]="-"
df_vfdic["Virulence_factor"]="-"
df_vfdic["fasID"]=df_vfdic.index
df_vfdic["tmp"].str.split("+",expand=True)
df_vfdic["Virulence_factor"]=df_vfdic["tmp"].str.split("+",expand=True)[0]
df_vfdic["VFID"]=df_vfdic["tmp"].str.split("+",expand=True)[1]
df_vfdic=df_vfdic.drop(["tmp"],axis=1,inplace=False)
df_vfdic
```

```python
#merge
df_cpr_hmmVFdb=pd.merge(df_VFs_meta,df_vfdic,left_on="VFID",right_on="VFID",how="outer")
df_cpr_hmmVFdb
```
