# Acquire DEG query

```python
import pandas as pd
import numpy as np
import sys, os
import re
```

处理DEG细菌

```python
#处理DEG10
#读入hmmscan原始文件
#sed '/^#/ d' ERR1191817.contigs_5M_contigs.hmmscan | sed 's/ \+/\\t/g' > 111.hmm.txt
input_DEG10="/lomi_home/gaoyang/software/CompRanking/databases/DEG/DEG10_Bacteria_tmp_hmm.txt"
df_DEG10=pd.read_csv(input_DEG10, 
                  sep=r'\\t', header=None, engine="python")
df_DEG10
#提取指定列
df_DEG10=df_DEG10.iloc[:,[0,2,4,5,7]]
df_DEG10.columns=["ID","Query","evalue","score","evalue_domain"]
#使用list方法修改query
df_DEG10=df_DEG10
empty=[]
for i in df_DEG10["Query"]:
    i=re.sub("_.*","",i)
    empty.append(i)
# empty_1=pd.Series(empty)
# df_test
df_DEG10["Query"]=empty
df_DEG10=df_DEG10[df_DEG10.evalue<1e-3]
df_DEG10=df_DEG10[df_DEG10.evalue_domain<1e-3]
df_DEG10
#获得Query
df_DEG10_Query=df_DEG10["Query"].unique()
df_DEG10_Query=pd.DataFrame(df_DEG10_Query)
# df_DEG10_Query.columns=["DEG_Query"]
df_DEG10_Query
```

处理DEG真菌

```python
#处理DEG20
#读入hmmscan原始文件
#sed '/^#/ d' ERR1191817.contigs_5M_contigs.hmmscan | sed 's/ \+/\\t/g' > 111.hmm.txt
input_DEG20="/lomi_home/gaoyang/software/CompRanking/databases/DEG/DEG20_Eukaryotes_tmp_hmm.txt"
df_DEG20=pd.read_csv(input_DEG20, 
                  sep=r'\\t', header=None, engine="python")
#提取指定列
df_DEG20=df_DEG20.iloc[:,[0,2,4,5,7]]
df_DEG20.columns=["ID","Query","evalue","score","evalue_domain"]
#使用list方法修改query
df_DEG20=df_DEG20
empty=[]
for i in df_DEG20["Query"]:
    i=re.sub("_.*","",i)
    empty.append(i)
# empty_1=pd.Series(empty)
# df_test
df_DEG20["Query"]=empty
df_DEG20=df_DEG20[df_DEG20.evalue<1e-3]
df_DEG20=df_DEG20[df_DEG20.evalue_domain<1e-3]
df_DEG20
#获得Query
df_DEG20_Query=df_DEG20["Query"].unique()
df_DEG20_Query=pd.DataFrame(df_DEG20_Query)
# df_DEG20_Query.columns=["DEG_Query"]
df_DEG20_Query
```

处理DEG古菌

```python
#处理DEG30
#读入hmmscan原始文件
#sed '/^#/ d' ERR1191817.contigs_5M_contigs.hmmscan | sed 's/ \+/\\t/g' > 111.hmm.txt
input_DEG30="/lomi_home/gaoyang/software/CompRanking/databases/DEG/DEG30_Archaea_tmp_hmm.txt"
df_DEG30=pd.read_csv(input_DEG30, 
                  sep=r'\\t', header=None, engine="python")
df_DEG30
#提取指定列
df_DEG30=df_DEG30.iloc[:,[0,2,4,5,7]]
df_DEG30.columns=["ID","Query","evalue","score","evalue_domain"]
#使用list方法修改query
df_DEG30=df_DEG30
empty=[]
for i in df_DEG30["Query"]:
    i=re.sub("_.*","",i)
    empty.append(i)
# empty_1=pd.Series(empty)
# df_test
df_DEG30["Query"]=empty
df_DEG30=df_DEG30[df_DEG30.evalue<1e-3]
df_DEG30=df_DEG30[df_DEG30.evalue_domain<1e-3]
df_DEG30
#获得Query
df_DEG30_Query=df_DEG30["Query"].unique()
df_DEG30_Query=pd.DataFrame(df_DEG30_Query)
# df_DEG10_Query.columns=["DEG_Query"]
df_DEG30_Query
```

合并DEG10,20,30

```python
#acquire non-duplicated Query of DEG data sets
df_concat_Query=pd.concat([df_DEG10_Query,df_DEG20_Query,df_DEG30_Query],ignore_index=True)
df_concat_Query=df_concat_Query.drop_duplicates()
df_concat_Query.columns=["DEG_Query"]
df_concat_Query
```

打开CompRanking_Virulence_Query.csv，本文件包含原始（没有任何去除）的VFDB对应的query

```python
input_cpr_vf="/lomi_home/gaoyang/software/CompRanking/databases/CompRanking_VirulenceDB/CompRanking_Virulence_Query2528.csv"
input_cpr_vf_withoutDomain="/lomi_home/gaoyang/software/CompRanking/databases/CompRanking_VirulenceDB/CompRanking_Virulence_Query_withoutDomain2623.csv"
df_cpr_vf=pd.read_csv(input_cpr_vf,
                  sep='\t', header=0, engine="python", index_col=0)
df_cpr_vf_withoutDomain=pd.read_csv(input_cpr_vf_withoutDomain,
                  sep='\t', header=0, engine="python", index_col=0)
df_cpr_vf
# df_cpr_vf_withoutDomain
```

将DEG——all和未清洗的原始hmm的query文件合并，准备剔除

```python
#merge cpr and DEG
df_crp_DEG=pd.merge(df_cpr_vf,df_concat_Query,left_on="Query",right_on="DEG_Query",how="left")
df_crp_DEG=df_crp_DEG.fillna("-")
df_crp_DEG
```
