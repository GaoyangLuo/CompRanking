```python
df_sum=pd.merge(df_AMR_MOB,df_VF_PATH,left_on="ORF_ID",right_on="ORF",how="left")
df_sum
```

```python
df_sum=df_sum.drop(["Contig_y","ORF","VF_Host_Bacteria","coverage","SNPs","mobileOG_ID","VF_Name"],axis=1,inplace=False)
df_sum
```

```python
df_sum.columns=["Contig","ORF_ID","ARG_prediction","ARG_class","ARG_Database","ARG_rank(recorded)","MGE_prediction","MGE_Name","MGE_Taxonomy","MGE_Category","MGE_Database","Mobility","Viculence_factor","VF_Category","VF_ESKAPE","Pathogenicity"]
df_sum["ARG_Rank(in-situ)"]="-"
df_sum=df_sum[["Contig","ORF_ID","ARG_prediction","ARG_class","ARG_Database","ARG_rank(recorded)","ARG_Rank(in-situ)","MGE_prediction","MGE_Name","MGE_Taxonomy","MGE_Category","MGE_Database","Mobility","Viculence_factor","VF_Category","VF_ESKAPE","Pathogenicity"]]
df_sum
```

```python
cpr_rank=[]
for i, name in df_sum.iterrows():
    if name["ARG_prediction"] == "-":
        cpr_rank.append("IV")
    elif name["ARG_prediction"]!= "-" and name["MGE_Name"] == "-":
        cpr_rank.append("III")
    elif name["ARG_prediction"]!= "-" and name["MGE_Name"] != "-" and name["Pathogenicity"] == "-":
        cpr_rank.append("II")
    elif name["ARG_prediction"]!= "-" and name["MGE_Name"] != "-" and name["Pathogenicity"] != "-":
        cpr_rank.append("I")
    else:
        cpr_rank.append("-")
df_sum["ARG_Rank(in-situ)"]=cpr_rank

df_sum
```
